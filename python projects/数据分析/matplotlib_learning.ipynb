```python
figure=plt.figure()
axes=plt.subplot(111,facecolor='y')
x=np.arange(-5,5,0.5)
y1=np.sin(x)
plt.plot(x,y1,ls=':',color='r',alpha=0.9,lw=1)
```

```python
def show_linestyles(x):
    style={
        '-':'-',
        '- -':'--',
        '-.':'-.',
        ':':':',
        'None':'None',
        ' ':' ', 
        '':'', 
        'solid': 'solid', 
        'dashed':'dashed',
        'dashdot':'dashdot',
        'dotted':'dotted'
    }
    padd=0
    for key,value in style.items():
        padd+=1
        y=np.sin(x+padd)
        plt.plot(x,y,ls=value,label=key,color=np.random.random(3))
    plt.legend()
x=np.linspace(0,2*np.pi,100)
plt.figure(figsize=(10,10))
show_linestyles(x)
```

```python
def show_drawstyles(x):
    style={
        'default':'default',
        'steps-mid':'steps-mid',
        'steps-pre':'steps-pre',
        'steps-post':'steps-post',
        'steps':'steps'
    }
    padd=0
    for key,value in style.items():
        padd+=1
        y=np.sin(x+padd)
        plt.plot(x,y,ds=value,label=key,color=np.random.random(3))
    plt.legend()
x=np.linspace(0,2*np.pi,100)
plt.figure(figsize=[10,10])
show_drawstyles(x)
```
