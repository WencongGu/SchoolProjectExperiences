## seaborn绘图操作
matplotlib是高度可定制的，负责处理绘图元素时比较方便，但是做出吸引人的细节就比较费力

seaborn负责图像的展示，自带定制主题和高级界面控制的matplotlib扩展包，基于matplotlib，相当于matplotlib的强化版，能让绘图变得更轻松

```python
import numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
%matplotlib inline

import seaborn as sns
sns.__version__
```

```python
x=np.arange(-5,5,0.1)
y1=np.sin(x)
plt.plot(x,y1)
y2=np.cos(x)
plt.plot(x,y2)
```

seaborn定制了几种较为美观的风格，通过sns.set()设置

二者都有一个参数，rc，是一个字典，包含的参数都是全局设置。通过语句sns.axes_style()获取其中可设置的参数

```python
x=np.arange(-5,5,0.1)
y1=np.sin(x)
plt.plot(x,y1)
y2=np.cos(x)
plt.plot(x,y2)
```

```python
def showstyles():
    x=np.arange(-np.pi,np.pi,0.1)
    plt.title('%colors')
    for i in range(6):
        plt.plot(x,np.sin(x+i))
showstyles()
```

```python
with sns.axes_style('darkgrid'):
    showstyles()
    plt.title('"with" attention')
```

```python
sns.axes_style()
```

```python
sns.set_style('darkgrid',rc={ 'axes.edgecolor': 'cyan'})
showstyles()
```

### 调色板
最主要的直接设置调色板的函数是color_palette()，提供了大部分的在seaborn中生成颜色的方式，并且可以用于任何函数内部的palette参数设置

color_palette()允许任意的seaborn调色板或matplotlib的色彩映射，还可以使用任何有效的matplotlib格式指定的颜色列表（rgb元组、十六进制颜色代码、HTML颜色名称）。返回值总是一个rgb元组的列表。

参数palette可选deep, muted, bright, pastel, dark, colorblind不同的风格，叫做分类调色板，默认10个颜色。

当分类调色板颜色不够用时，圆形调色板可以提供更多的色彩。圆形调色板中比较常用的就是hls色彩空间。设置参数palette为"hls"即可。

参数n_colors设置颜色数量，在默认的颜色区间内取值

在线性调色板中有参数as_cmap使得颜色变得更加连续

seaborn专门展示颜色的函数palplot([(r,g,b),(r,g,b),...]

```python
sns.color_palette()
```

```python
colors=sns.color_palette()
sns.palplot(colors)
```

```python
colors=sns.color_palette(palette='muted')
sns.palplot(colors)
```

```python
colors=sns.color_palette(palette='hls')
sns.palplot(colors)
```

```python
colors=sns.color_palette(palette='hls',n_colors=20)
sns.palplot(colors)
```

```python
colors=sns.color_palette(palette='hls',n_colors=20)
sns.palplot(colors)
```

#### 颜色微调
函数hls_palette()/husl_palette()

控制颜色的亮度lightness/l，饱和度saturation/s，起始颜色值h

```python
colors=sns.color_palette(palette='hls',n_colors=8)
sns.hls_palette(s=0.75)
sns.palplot(colors)
sns.hls_palette(s=0.55)
sns.palplot(colors)
```

#### husl颜色
由于人眼的工作特性，在rgb度量上一致的颜色在视觉上并不平衡，比如黄色绿色较亮，蓝色较暗。

为了与hls系统一致，seaborn为husl系统提供了一个接口，使得选择均匀间隙的颜色更加便捷，同时保持亮度和饱和度更加一致

```python
colors=sns.color_palette(palette='hls')
sns.palplot(colors)
colors=sns.color_palette(palette='husl')
sns.palplot(colors)
```

#### 调色板工具choose_colorbrewer_palette()
需要传递一个参数data_type，可选'sequential'连续调色（渐变色）, 'diverging'两级调色（两级分化）, 'qualitative'分类调色

另一种工具是choose_dark_palette(),choose_diversing_palette(),

使用一个变量接收调色结果，自动保存

```python
colors=sns.choose_colorbrewer_palette(data_type='sequential')
```

```python
sns.palplot(colors)
```

```python
sns.choose_dark_palette(input='hls')
```

```python
colors=sns.choose_colorbrewer_palette(data_type='sequential',as_cmap=True)
```

### seaborn 常用绘图
seaborn相对于matplotlib最大的不同就是自带了分组功能，而matplotlib需要先对数据进行处理

```python
# 使用sns.load_dataset()加载数据集，但是要用外网下载，所以不行
tips=sns.load_dataset('titanic')
tips
```

#### 普通2D图像

##### 通用参数
seaborn自动识别分组or聚合

重要的参数：

x，y：要展示的数据，seaborn绘图时会自动对x进行分组，得到几组数据，默认分别计平均值并展示

data：数据来源，通常是一个dataframe

hue：分组因子，用于映射颜色的列。在对x分组之后再进一步分组，组内的每个组颜色用调色板中的颜色标注

palette：用于颜色映射的调色板，注意调色板要使用sns.color_palette(colors)生成调色板对象

order：排序的序列，如order=['dinner','lunch']，操作x的分组展示顺序

hue_order：颜色映射的序列，例如hue order=['setosa','virginica']，展示x的分组再根据hue分组的展示顺序

##### 条形图
sns.barplot()

默认展示的是mean，通过参数estimator设置函数名调整

置信区间的概率默认是95%，通过参数ci调整，设置为None时不显示

##### 计数条形图
sns.countplot()

单变量分析，统计每一类数据的个数，设定了x或者y后另一个轴就固定是count

##### 散点图
当然不需要分组或者聚合

sns.stripplot()

注意它会将x轴数据当作离散值（字符串）处理，而且显示的内容无法更改，因此更适合显示离散值和连续值的对应关系。如果x轴数据太多（连续值），会密密麻麻全显示出来。显示连续值与连续值的关系使用scatterplot

sns.scatterplot()

##### 分族散点图
sns.swarmplot()

显示离散值与连续值的关系，但是做了优化，使点不会重叠，更清晰地显示数据分布

##### 小提琴图
sns.violinplot()

和分族散点图一样，但是进一步优化，显示出来不是点，而是连续的面，像小提琴一样。

设置参数hue后会分裂，设置split为True后合并

##### 箱线图
sns.boxplot()

参数：

linewidth设置线宽

orient='h'方向

whis是否显示离群点，为np.inf时不显示

##### 热图
sns.heatmap()

使用函数data.corr()计算出data各列两两的相关系数，得到一个DataFrame。参数method可选pearson（连续性数据之间且服从正态分布）、kendall（打分型数据等离散型）、spearman（连续性且不服从正态分布）。数据跨度太大时会影响计算准确性，可以不去关注具体大小，只关注大小关系并加以处理

sns.heatmap()参数：

cmap有此参数（相当于colormap）就是设置颜色映射的。annot为True时会显示数值

#### 高级函数
seaborn与DataFrame结合得非常好，因此传参可以直接传列名，返回值是FacetGrid（平面网格）对象

主要参数：

x，y：变量

data：数据源

hue：进阶分组依据变量

style；设置后可以让每组数据的绘图风格不一致

col、row：确定平面网格图布局的变量

kind：底层作图类型名称，如line，bar，box

##### 可视化变量关系（relationship）
会在图像中加上一条回归曲线

regplot()观察数据间关系，默认散点图

lmplot()中x，y必须以字符串的方式传入，比regplot更加严格，但是作用都一样

参数：

x_jitter，y_jitter：仅适用于离散型数据，给离散型数据添加抖动，且不影响拟合回归线本身

x_estimator：当x轴是离散型数据时，指定y轴上数据的运算函数，如np.mean，np.max

order：如果数据的分布规律是非线性的，任意设置一个大于1的数，则使用多项式回归来绘制回归曲线

##### 单变量分布
直方图sns.dist()

参数：

kde控制是否作核密度估计

density控制是否显示频率

bins区间个数

rug为True时会在直方图下面显示出小竖线，每条线代表了一个数据，可以更细致地反映数据分布（地毯图）

其他形式如xxx_kws的参数输入字典，控制相应主体的性质

sns.kdeplot()函数绘制kde图，而且可以绘制两组数据的kde曲线，横轴是data，纵轴是data2

sns.rugplot()地毯图

##### 双变量分布
sns.joinplot()相当于散点图的加强版，在散点图的上轴和下轴分别显示了两个变量的具体直方图形式的分布

参数：

xlim，ylim

space：主画布和分画布的距离（0~1）

marginal_kws：分支画布外观属性字典

joint_kws：主画布（散点图）外观属性，散点图的各种属性都可以设置

kind：设置主画布绘图类型，默认散点图，选择其他类型要保证表示二者关系的图像可以画出来

##### 成对变量分布
sns.pairplot()类似pandas绘图中的散点图矩阵，pandas中用的是语句d.plotting.scatter_matrix()，得到两两之间的关系散点图。也可以分组

参数

var、x_var、y_var：要显示的列

markers：列表，分组的点型

#### 多层面板分类图
sns.factorplot()可以使用更多的子视图来显示更多的数据关系

也可以使用FacetGrid()对象

实操：

创建FacetGrid对象

fg=sns.FacetGrid(data=data,row='row_name',col='column_name',...)

在每个分区绘图

fg.map(func=plt.xxx,'inner_row','inner_column')

注意参数，func必须是matplotlib中的绘图函数，一旦指定，绘图时会依次传入后面的'inner_row','inner_column'作为绘图数据，因此后面的参数必须与之匹配
