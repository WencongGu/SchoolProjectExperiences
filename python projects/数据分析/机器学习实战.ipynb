支持部分markdown语法

基础：

\# 用于标题，注意和后面文字间加一个空格，一共六级标题，按 \# 的数量区分

\$ 用于公式输入，一对\$~\$用于普通公式如$\frac{d\theta}{dt}=\omega$，两对\$\$~\$\$用于行间公式，居中，妙用：标题居中

\* 用于斜体、加粗，一对\*~\*用于*斜体*，两对\*\*～\*\*用于**加粗**，三对\*\*\*～\*\*\*用于***加粗且斜体***

# 一级标题
## 二级标题
### 三级标题
#### 四级标题
##### 五级标题
###### 六级标题
*这是斜体*

_这是斜体_

**这是粗体**

__这是粗体__

***这是粗斜体***

___这是粗斜体___

换行：

1. 直接在一句话后敲两个空格  
1. 两句话之间加一个空行

1. 如果你在编辑的时候，想让一行文字在显示的时候换行，就在中间加<br/>就可以换行了

引用：

>这是一个引用
>>这是一个引用的引用
>>>这是一个引用的引用的引用

插入链接：

[链接名称](链接地址)

或者

<httpe://www.baidu.com>

插入图片：

![图片描述，可写可不写，但是中括号要有](图片地址，本地链接或者URL地址。)

也可以直接复制粘贴

表格:

使用`|`分割不同的单元格  
使用`-`来分隔表头和其他行  
使用`:-`将表头及单元格内容左对齐  
使用`-:`将表头及单元格内容右对齐  
使用`:-:`将表头及单元格内容居中

|项目|价格|数量|

|计算机|\$1600|5|
|手机|\$12|12|
